Mimicing the carnival guessing game named "Three cup monte" using methods and functions with the help of lists.


```python
#creating a simple list
from random import shuffle
my_list=[' ','O',' ']
```

```python
#shuffling the list


def shuffle_list(my_list):
    shuffle(my_list)
    return my_list
```

```python


shuffle_list(my_list)
```

```python
#function for the player to guess

def user_guess():
    guess = ''
    
    while guess not in ['0','1','2']:
        guess = input("Pick a number: 0, 1, or 2:  ")
    
    return int(guess)    
    
```

```python
user_guess()
```

```python
#checking the guess is true or not
def check_guess(my_list,guess):
    if my_list[guess]=='O':
        print("Correct!")
    else :
        print("Wrong guess,Try again!")
        print(my_list) 
```

```python


# Initial List
mylist = [' ','O',' ']

# Shuffling
mixedup_list = shuffle_list(mylist)

# Get User's Guess
guess = user_guess()

# Check User's Guess
check_guess(mixedup_list,guess)
```
